## 03 Visualizations

```python
# Bar Chart for orders_days_of_week
df_merged['orders_day_of_week'].value_counts().plot.bar()
```

```python
# Bar Chart for orders_days_of_week (sort by index)
df_merged['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

## 04 Sampling Data

```python
# random sample .7
np.random.seed(4)
dev = np.random.rand(len(df_merged) <= 0.7)
```

```python
# create a list holding true/false values to the test np.random.rant() < = 0.7
dev = np.random.rand(len(df_merged)) <= 0.7
```

```python
# store 70% of the sample in the dataframe â€œbigâ€
big = df_merged[dev]
```

```python
# store 30% of the sample in the dataframe â€œsmallâ€
small = df_merged[~dev]
```

```python
# check
len(df_merged)
```

```python
len(big) + len(small)
```

```python
df_2 = small[['orders_day_of_week','prices']]
```

```python
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```
